#### 2.4.1 无穷小运算法则

两个无穷小的和是无穷小

两个无穷小的差是无穷小

两个无穷小的乘积是无穷小

有限个无穷小的和、差与积都是无穷小

有界函数与无穷小的乘积是无穷小


#### 2.4.2 极限运算法则

例2.15 求$$\lim_{x\rightarrow1}\frac{x-1}{x^2-1}$$

解：

$$\lim_{x\rightarrow1}\frac{x-1}{x^2-1} = \lim_{x\rightarrow1}\frac{x-1}{(x+1)(x-1)}=\lim_{x\rightarrow1}\frac{1}{x+1}=\frac{1}{2}$$



例2.16 求$$\lim_{x\rightarrow1}\frac{x+2}{x^2-1}$$

解：因为$$\lim_{x\rightarrow1}\frac{x^2-1}{x+2}=0$$

所以为$$\lim_{x\rightarrow1}\frac{x+2}{x^2-1}=\infty$$

#### 2.7.2 函数的间断点 

函数f(x)在点$x_0$处不连续的三种情形：

(1)在x=$x_0$处没有定义

(2)在x=$x_0$处有定义，但$\lim_{x\rightarrow{x_0}}f(x)$不存在

(3)在x=$x_0$处有定义，$\lim_{x\rightarrow{x_0}}f(x)$存在，但$\lim_{x\rightarrow{x_0}}f(x)\neq{f(x_0)}$

#### 2.7.1 函数的连续性

有$\lim_{\Delta{x}\rightarrow0}\Delta{y}=0$，则函数y=f(x)在点$x_0$处连续

等同于：

有$\lim_{\Delta{x}\rightarrow{x_0}}f(x)=f(x_0)$，则函数y=f(x)在点$x_0$处连续

#### 2.8.4 初等函数的连续性

例2.38 求$$\lim_{x\rightarrow0}\frac{\sqrt{1+x^2}-1}{x^2}$$

解：$$\lim_{x\rightarrow0}\frac{\sqrt{1+x^2}-1}{x^2}=\lim_{x\rightarrow0}\frac{(\sqrt{1+x^2}-1)(\sqrt{1+x^2}+1)}{x^2(\sqrt{1+x^2}+1)}=\lim_{x\rightarrow0}\frac{1}{\sqrt{1+x^2}+1}=\frac{1}{2}$$
