#### 3.4.1隐函数的导数

显函数：函数y=f(x)称为**显函数**，因为因变量y直接用自变量x的一个式子表示了出来，y与x之间的函数关系很明显。

隐函数：有时因变量y与自变量x之间的对应关系没有用公式y=f(x)明显给出，或不能明显给出，而是用x，y之间的一个方程式F(x,y)=0来表示的，这种由方程F(x,y)=0所确定的函数称为**隐函数**。

隐函数的显化：从方程F(x,y)=0中有时可以解出y来，这时便得到了显函数，这叫做**隐函数的显化**。例如可以从隐函数方程3x+5y+1=0解出显函数$y=-\frac{3}{5}x-\frac{1}{5}$

但有时隐函数并不能表示为显函数的形式，或者表示出来很复杂，那么在不解出y的情况下，怎么求导数y'?下面举例解决这个问题。

例3.20 求由方程$e^y+xy-e=0$所确定的隐函数的导数$\frac{dy}{dx}$

解：

$$e^y\frac{dy}{dx}+y+x\frac{dy}{dx}=0$$

$$\frac{dy}{dx}=-\frac{y}{e^y+x}$$

求隐函数$x^3+y^3=xy$的导数y'

解：

$$3x^2+3y^2\frac{dy}{dx}=y+x\frac{dy}{dx}$$

$$(3x^2-x)\frac{dy}{dx}=y-3x^2$$

$$\frac{dy}{dx}=\frac{y-3x^2}{3x^2-x}$$

$$x'=\frac{dx}{dy}=\frac{3x^2-x}{y-3x^2}$$

例3.21 求由方程$y^5+2y-x-3x^7=0$所确定的隐函数在x=0处的导数$\frac{dy}{dx}|_{x=0}$

解：

$$5y^4\frac{dy}{dx}+2\frac{dy}{dx}-1-21x^6=0$$

$$\frac{dy}{dx}=\frac{21x^6+1}{5y^4+2}$$

由原方程知道，x=0,

$$y^5+2y=0$$
$$y(y^4+2)=0$$
$$y=0$$

代入导数函数，$$\frac{dy}{dx}|_{x=0}=\frac{1}{2}$$

例3.23 求由方程$x-y+\frac{1}{2}siny=0$所确定的隐函数的二阶导数$\frac{d^2y}{dx^2}$

$$=-\frac{y-x\frac{dy}{dx}}{y^2}$$

$$=-\frac{r^2}{y^3}$$
